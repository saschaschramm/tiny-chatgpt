```python
from chatgpt.mdp import MDP
from chatgpt.mdp_nodes import StateNode
import mdpviz

root: StateNode = StateNode(identifier=1, timesteps=2, t=0, parent=None)
mdp: MDP = MDP(root)
mdp.init()
mdp.add_random_policy()
mdpviz.draw(mdp, path=None)
```

### Explanation
* In case of RLHF the language model is an agent.
* For simplicity we assume that the answer has a length of `2` words (number of timesteps is `2`).
* t=0: The language model takes a question `S0=1` as input and generates the first word of the answer e.g. `A0=0` which leads to the next state `S1=10`
* t=1: The language model takes `S1=10` as input and generates the second word of the answer `A1=0` which leads to the next state `S2=100`
* t=2: If `S2` contains an even number of 1s then the environment returns a reward of `1`.

### Goal
The goal of the agent is to maximize the discounted reward.

### Notation
There are different notations for the reward:
* PPO: Transitioning from `S0` to `S1` leads to a reward of `R0`.
* Sutton et al.: Transitioning from `S0` to `S1` leads to a reward of `R1`.

We use the notation of Sutton et al.
