```python
from chatgpt.mdp import MDP
from chatgpt.mdp_nodes import StateNode
import mdpviz
```

```python
# Generalized advantage estimation

# A MDP with two timesteps
root: StateNode = StateNode(identifier=1, timesteps=2, t=0, parent=None)
mdp: MDP = MDP(root, gamma=0.9, lam=1.0)

# Add transition probabilities and rewards
mdp.init()

# Add a deterministic policy that always takes action `1`.
mdp.add_deterministic_policy()

# k=0: Compute the generalized advantage estimate for each state.
mdp.gae()

# k=0: Using the generalized advantage estimate, the state values are updated.
mdp.update_state_values(learning_rate=0.1)

# k=1: Compute the generalized advantage estimate for each state.
mdp.gae()

# Draw the MDP after 2 iterations.
mdpviz.draw(mdp, path=None)
```

Generalized advantage estimation is equivalent to n-step temporal difference learning when lambda is set to `1`.

```python
# Generalized advantage estimation

# A MDP with two timesteps
root: StateNode = StateNode(identifier=1, timesteps=2, t=0, parent=None)
mdp: MDP = MDP(root)

# Add transition probabilities and rewards
mdp.init()

# Add a deterministic policy that always takes action `1`.
mdp.add_deterministic_policy()

# k=0: Compute the temporal difference error for each state.
mdp.temporal_difference(gamma=0.9)

# k=0: Update the state values using the temporal difference error.
mdp.update_state_values(learning_rate=0.1)
mdpviz.draw(mdp, path=None)

# k=1: Compute the temporal difference error for each state.
mdp.temporal_difference(gamma=0.9)

# Draw the MDP after 2 iterations.
mdpviz.draw(mdp, path=None)
```

Looking at the two plots, we can see that the state values are the same for both methods.
